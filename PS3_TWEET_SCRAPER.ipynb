```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import datetime
import snscrape.modules.twitter as sntwitter
import itertools
import csv 



# Setting variables to be used below
maxTweets = 50000

# Creating list to append tweet data to
tweets_list2 = []
# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('covid19 since:2020-05-04 until:2020-09-30').get_items()):
    if i < maxTweets:
        tweets_list2.append([tweet.id, tweet.content])
    else: break
# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['Tweet Id', 'Text'])

#writing data in csv file
tweets_df2.to_csv('text-tweets.csv', encoding='utf-8')

display(tweets_df2)

tweets_df2.shape

```
